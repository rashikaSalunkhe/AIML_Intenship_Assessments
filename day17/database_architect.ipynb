```python
import sqlite3
```

```python
conn = sqlite3.connect("internship.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS interns (
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    track TEXT NOT NULL,
    stipend INTEGER
)
""")
```

```python
cursor.executemany("""
INSERT INTO interns (id, name, track, stipend)
VALUES (?, ?, ?, ?)
""", [
    (1, 'Aarav', 'Data Science', 15000),
    (2, 'Riya', 'Web Development', 12000),
    (3, 'Kabir', 'Data Science', 18000),
    (4, 'Meera', 'Cyber Security', 20000),
    (5, 'Arjun', 'Web Development', 14000)
])
```

```python
conn.commit()
```

```python
cursor.execute("SELECT name, track FROM interns")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
conn.close()
```
