```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Marks": [45, 50, 52, 48, 47, 49, 51, 300]  # 300 is extreme value
}

df = pd.DataFrame(data)
df
```

```python
mean = df["Marks"].mean()
std = df["Marks"].std()

print("Mean:", mean)
print("Standard Deviation:", std)
```

```python
df["Z_score"] = (df["Marks"] - mean) / std
df
```
