```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\VICTUS\Downloads\housing.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
plt.figure(figsize=(8,5))
```

```python
sns.scatterplot(x='median_income', y='median_house_value', data=data)
plt.title("Median Income vs Median House Value")
plt.xlabel("Median Income")
plt.ylabel("Median House Value")

plt.show()
```

```python
correlation = data[['median_income', 'median_house_value']].corr()

print(correlation)
```

```python
plt.figure(figsize=(8,5))
```

```python
sns.boxplot(x='ocean_proximity', y='median_house_value', data=data)

plt.title("House Value by Ocean Proximity")
plt.xticks(rotation=45)

plt.show()
```
