```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv(r"C:\Users\VICTUS\Downloads\housing.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['median_house_value'], kde=True)
plt.title("Distribution of Median House Value")
plt.xlabel("Median House Value")
plt.ylabel("Frequency")
plt.show()
```

```python
skewness = data['median_house_value'].skew()
kurtosis = data['median_house_value'].kurt()
print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='ocean_proximity', data=data)

plt.title("Count of Houses by Ocean Proximity")
plt.xticks(rotation=45)

plt.show()
```

### Interpretation:

The histogram shows that the median house value distribution is positively skewed.
The skewness value confirms that the distribution is right-skewed.
This indicates that most houses are in lower price range, with few very high-priced houses.

The count plot shows that most houses belong to the INLAND category.
