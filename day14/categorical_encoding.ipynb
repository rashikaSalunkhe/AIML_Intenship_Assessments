```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data = {
    'Transmission': ['Automatic', 'Manual', 'Automatic', 'Manual', 'Automatic'],
    'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue']
}

df = pd.DataFrame(data)

df
```

## Step 1: Apply Label Encoding to Transmission
Transmission is binary, so Label Encoding is appropriate.


```python
le = LabelEncoder()

df['Transmission'] = le.fit_transform(df['Transmission'])

df
```

## Step 2: Apply One-Hot Encoding to Color
Color is nominal (no order), so One-Hot Encoding is used.
To avoid Dummy Variable Trap, drop_first=True is applied.


```python
df = pd.get_dummies(df, columns=['Color'], drop_first=True)

df
```
